```python
df.drop(columns=['Remarks'], inplace=True)
df
```

```python
df.set_index('Sr No', inplace=True)
```

```python
df.info()
```

```python
df.rename(columns = {
    'Date dd/mm/yyyy': 'date',
    'Startup Name': 'startup',
    'Industry Vertical': 'vertical',
    'SubVertical': 'subvertical',
    'City  Location': 'city',
    'Investors Name': 'investors',
    'InvestmentnType': 'round',
    'Amount in USD': 'amount'
}, inplace=True)
df
```

```python
df.info()
```

```python
df['amount'] = df['amount'].fillna('0')
```

```python
df.info()
```

```python
df['amount'] = df['amount'].str.replace(',', '')
df['amount'] = df['amount'].str.replace('undisclosed', '0')
df['amount'] = df['amount'].str.replace('unknown', '0')
df['amount'] = df['amount'].str.replace('Undisclosed', '0')
df['amount'] = df['amount'].str.replace('[^0-9.]', '', regex=True)
# df['amount'] = df[df['amount'].str.isdigit()] this also works but removes
df['amount'] = df['amount'].astype('float')
# df['amount'].astype('int')
```

```python
df.info()
```

```python
df.head()
```

```python
def to_npr(dollar):
    npr = dollar * 134;
    return npr/10000000
```

```python
df['amount'] = df['amount'].apply(to_npr)
```

```python
df.head()
```

```python
# df['date'] = df['date'].replace('05/072018', '05/07/2018')
# df['date'] = df['date'].replace('01/07/015', '01/07/2015')
# df['date'] = df['date'].replace(r'\\xc2\\xa0', '', regex=True)
df['date'] = pd.to_datetime(df['date'])
df.head()
df.info()
```

```python
df = df.dropna(subset=['date', 'startup', 'vertical', 'city', 'investors', 'round', 'amount'])
```

```python
df.info()
```

```python
df.to_csv('startup_cleaned.csv', index=False)
```
